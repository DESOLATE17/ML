Для заданного набора данных произведите масштабирование данных (для одного признака) и преобразование категориальных признаков в количественные двумя способами (label encoding, one hot encoding) для одного признака. Какие методы Вы использовали для решения задачи и почему?

Для решения задачи масштабирования данных и преобразования категориальных признаков в количественные с использованием датасета sklearn.datasets.load_iris, я использовала следующие методы:

**Масштабирование данных**: Для масштабирования данных я использовала метод MinMaxScaler из библиотеки sklearn.preprocessing. Этот метод преобразует данные так, чтобы минимальное значение каждого признака было равно 0, а максимальное значение — 1. Это полезно для алгоритмов машинного обучения, которые могут быть чувствительны к масштабу признаков.

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.datasets import load_iris
import pandas as pd

iris = load_iris()
iris_df = pd.DataFrame(iris.data, columns=iris.feature_names)
iris

minmax = MinMaxScaler()
minmax_iris = minmax.fit_transform(iris_df)
minmax_iris = pd.DataFrame(minmax_iris, columns=iris.feature_names)
```

**Преобразование категориальных признаков в количественные**

**One-Hot Encoding**: Для преобразования категориальных признаков в количественные я использовала OneHotEncoder из sklearn.preprocessing. Этот метод создает новые бинарные признаки для каждой категории в исходном признаке.

```python
from sklearn.preprocessing import OneHotEncoder

iris_df['species'] = iris.target
one_hot_encoder = OneHotEncoder(sparse=False)
one_hot_encoded = one_hot_encoder.fit_transform(iris_df[['species']])
one_hot_df = pd.DataFrame(one_hot_encoded, columns=one_hot_encoder.get_feature_names_out())
one_hot_df
```

**Label Encoding**: Для преобразования категориальных признаков в количественные я использовала LabelEncoder из sklearn.preprocessing. Этот метод присваивает каждой категории уникальное числовое значение. Но в данном датасете изначально категориальный признак представлен в виде числа в колонке species.

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
iris_df['species_encoded'] = label_encoder.fit_transform(iris_df['species'])
iris_df
```
